## Importando dados

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/points_tmw.csv')
df.head()
```

## Frequencia Absoluta

É quantas vezes um valor aparece.

fᵢ = número de ocorrências do valor xᵢ


```python
freq_produto = (df.groupby(['descProduto'])[['idTransacao']].count())
```

## Frequencia Absoluta Acumulada

É a soma progressiva das frequências absolutas.

Fᵢ = f₁ + f₂ + f₃ + ... + fᵢ

```python
freq_produto['freq_abs_acum'] = freq_produto['idTransacao'].cumsum()
```

## Frequencia Relativa

É a proporção de cada valor em relação ao total:

frᵢ = fᵢ / n

```python
freq_produto['freq_rel'] = freq_produto['idTransacao'] / freq_produto['idTransacao'].sum()
```

## Frequencia Relativa Acumulada

Proporção de cada valor em relação ao total de dados:

frᵢ(%) = (fᵢ / n) × 100

```python
freq_produto['freq_rel_acum'] = freq_produto['freq_rel'].cumsum()
```

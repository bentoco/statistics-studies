```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/points_tmw.csv')
df
```

```python
freq_categ = (df.groupby(['descCategoriaProduto'])[['idTransacao']]
                .count()
                .rename(columns={'idTransacao': 'freq_abs'}))

freq_categ['freq_abs_acum'] = freq_categ['freq_abs'].cumsum()

freq_categ['freq_rel'] = freq_categ['freq_abs'] / freq_categ['freq_abs'].sum()
freq_categ['freq_rel_acum'] = freq_categ['freq_rel'].cumsum()

display(freq_categ)
```
